```python
import requests
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
import re
```

```python
sitemaps = pd.read_xml("https://datanews.levif.be/news-sitemap.xml")
sitemaps.head()
```

```python
urls = sitemaps["loc"]
print(urls)
```

```python
url = urls[0]
response = requests.get(url).text
print(response)
```

```python
soup = BeautifulSoup(response, "html.parser")
print(soup)
```

```python
def find_my_shit(url):
    title = soup.find_all("h1")
    article_title = title[0].text.strip()
    return article_title


print(find_my_shit(url))
```

```python
article = soup.find("p").text
print(article)
```

```python
script = soup.find("script", {"type": "application/ld+json"})
data = json.loads(script.text, strict=False)

accessing_list = data["@graph"]
accessing_dict = accessing_list[0]
published_date = accessing_dict["datePublished"]
print(published_date)
```
