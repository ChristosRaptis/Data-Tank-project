```python
import requests
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
import re
```

```python
sitemaps = pd.read_xml("https://datanews.levif.be/ict/actualite-recente/sitemap-news.xml")
sitemaps.head()
```

```python
urls = sitemaps["loc"]
print(urls)
```

```python
url = urls[0]
response = requests.get(url).text
print(response)
```

```python
soup = BeautifulSoup(response, "html.parser")
print(soup)
```

```python
def find_my_shit(url):
    title = soup.find_all("h1")
    article_title = title[0].text.strip()
    return article_title


print(find_my_shit(url))
```
