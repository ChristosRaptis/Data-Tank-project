```python
def find_article_text(url: str) -> str:
    response = requests.get(url)
    soup = bs(response.content, "html.parser")
    paragraphs = [p.text for p in soup.find_all("p", attrs={"class": None})]
    article_text = "".join(paragraphs)
    return article_text
```
