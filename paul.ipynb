```python
import requests
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# create csv, with url, title, text and date for each article
sitemaps = pd.read_xml("https://www.rtbf.be/site-map/articles.xml")
sitemaps.head()
```

```python
urls = sitemaps["loc"]
print(urls)
```

```python
test = urls[1]
response = requests.get(test).text
print(response)
```

```python
soup = BeautifulSoup(response, "html.parser")
print(soup)
```

```python
title = soup.find("h1").text
print(title)
```

```python
article = soup.find_all("p", attrs={"class": None})
articles = [p.text for p in article]
cleaned_articles = " ".join(articles)
print(cleaned_articles)
```
