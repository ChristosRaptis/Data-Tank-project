```python
import requests
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# create csv, with url, title, text and date for each article
sitemaps = pd.read_xml("https://www.rtbf.be/site-map/articles.xml")
sitemaps.head()
```

```python
urls = sitemaps["loc"]
print(urls)
```

```python
test = urls[1]
response = requests.get(test).text
print(response)
```

```python
headline = BeautifulSoup(response, "html.parser")
# print(headline.find("html", {"html dir": "headline"}))
print(list(headline.children))
```
