```python
import requests
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
# create csv, with url, title, text and date for each article
sitemaps = pd.read_xml("https://www.vrt.be/vrtnws/nl.news-sitemap.xml")
sitemaps.head()
```

```python
urls = sitemaps["loc"]
print(urls)
```

```python
test = urls[1]
split = test.split("/")
date = split[5], split[6], split[7]
media = split[2]
print(date)
print(media)
```

```python
url = urls[2]
response = requests.get(url).text
print(response)
```

```python
soup = BeautifulSoup(response, "html.parser")
print(soup)
```

```python
title = soup.find("h1").text
print(title)
```

```python
article = soup.find_all("p", attrs={"class": None})
articles = [p.text for p in article]
cleaned_articles = " ".join(articles)
print(cleaned_articles)
```
