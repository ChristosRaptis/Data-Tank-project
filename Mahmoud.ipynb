```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.demorgen.be/sitemaps/news.xml'
page_divs = requests.get(url).content
soup = BeautifulSoup(page_divs, "html.parser")
print(soup.prettify())
```

```python
data = {'lastmod':[],'url-parent':[]}
all_tables = soup.find_all(["loc",'lastmod'])
for i in range(1, len(all_tables) - 1,2):
    print(all_tables[i].text)
    data['url-parent'].append(all_tables[i+1].text)
    data['lastmod'].append(all_tables[i].text)
```

```python
print(data)
```

```python
pd.DataFrame.from_dict(data)
```

```python

link_divs = requests.get(data['url-parent'][0]).content
soup_link = BeautifulSoup(link_divs, "html.parser")
print(soup_link.prettify())
```
