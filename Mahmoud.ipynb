```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.demorgen.be/sitemap.xml'
page_divs = requests.get(url).content
soup = BeautifulSoup(page_divs, "html.parser")
print(soup.prettify())
```

```python
data = {'date':[],'url':[],'title':[],'text':[]}
all_tables = soup.find_all(["loc",'lastmod'])
for i in range(1, len(all_tables) - 1,2):
    print(all_tables[i].text)
    data['url'].append(all_tables[i].text)
    data['date'].append(all_tables[i+1].text)
    data['title'].append('')
    data['text'].append('')
```

```python
print(data)
```

```python
pd.DataFrame.from_dict(data)
```
