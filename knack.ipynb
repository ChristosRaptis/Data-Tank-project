```python
import requests
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
sitemaps = pd.read_xml("https://www.knack.be/news-sitemap.xml")
sitemaps.head()
```

```python
urls = sitemaps["loc"]
print(urls)
```

```python
url = urls[0]
response = requests.get(url).text
print(response)
```

```python
soup = BeautifulSoup(response, "html.parser")
print(soup)
```

```python
title = soup.find("h1").text
print(title)
```

```python
article = soup.find_all("p", attrs={"class": None})
articles = [p.text for p in article]
cleaned_articles = " ".join(articles)
print(cleaned_articles)
```

```python
date = soup.find_all(
    '<meta property="article:published_time" content="2023-08-01Tui:04:25+00:00" />'
)
print(date)
```
